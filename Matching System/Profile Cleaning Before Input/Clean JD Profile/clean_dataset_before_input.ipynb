```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
import spacy

nlp = spacy.load("en_core_web_sm")
lemmatizer = nlp.get_pipe("lemmatizer")

class clean_original_dataset():

    cleaned_info = {}

    def __init__(self,data,company,position,id):
        # data should be in a dictionary format (refer to the example above)
        '''# TODO:(1) convert all company names, positions and ids in this class and original dataset to lower cases.'''
        self.data = data
        self.company = company
        self.position = position
        self.id = id

    @ staticmethod
    def get_degree_info(s):
        if s == "None":
            return []
        else:
            degree = []
            '''# TODO: (2) this classification system should be double checked and extended'''
            degree_info={"primary school":1,
            "special education":2,
            "ged":3,
            "general education":3,
            "secondary":4,
            "junior high school":5,
            "high school":6,
            "certification":7,
            "vocational":8,
            "college":9,
            "HND":10,
            "HNC":10,
            "higher national":10,
            "associate":11,
            "international":12,
            "bs":13,
            "ba":13,
            "bachelor":13,
            "ms":14,
            "ma":14,
            "master":14,
            "mba":14,
            "intermediate graduate":15,
            "professional":15,
            "post professional":16,
            "doctor":17,
            "phd":17,
            "md":17,
            "post doctor":18}
            for degree_level in degree_info.keys():
                if degree_level in s:
                    degree.append(degree_info[degree_level])
                else:
                    continue
            degree = list(set(degree))
            return degree

    @ staticmethod
    def get_major_info(s):
        if s == "None":
            return []
        else:
            major = []
            ''' # TODO: (3) Can label each major manually to find related majors. - give each major multiple labels!'''
            ''' example: computer science - computer - science - technical'''
            df_subject = pd.read_excel("temp_merged_major.xlsx")
            subject_list = list(df_subject['major'])
            # if exists, find it out directly
            for subject in subject_list:
                if subject in s:
                    major.append(subject)
                else:
                    continue

            # if not exist, find all related subjects
            for remaning_word in s.split():
                for i in subject_list:
                    if remaning_word in i and remaning_word!= "science" and remaning_word!= "related": # could extend unrelated words, but it would be better if we can use labels!!!
                        major.append(i)
                    else:
                        continue
            major = list(set(major))
            return major

    @ staticmethod
    def get_yoe_info(s):
        # Question: if need to be more concise?
        if s == "None":
            return [0,""]
        else:
            work_exp = [str(i) for i in range(11)]
            require_year = 0
            for i in work_exp:
                if i in s:
                    require_year = int(i)
                    break
            s_new = re.sub('[^a-zA-Z]'," ",s)
            s_new = s_new.replace("years","").replace("year","").replace("experiences","").replace("experience","")
            new_list = s_new.split()
            s_new = " ".join(new_list)
            return [require_year, s_new]
    
    def clean_degree_and_yoe(self):
        s_degree = self.data[self.company][self.position][self.id]['Degree&YOE_Requirement']
        final_output = {}
        if s_degree == "None|None|None|None":
            return {}
        else:
            # split the sentences and store in the list (can be used for deducting the classification of words)
            sentences = s_degree.split("|")
            # store cleaned sentences
            corpus = []
            # clean data
            for i in range(len(sentences)):
                review = sentences[i].replace("etc.","").replace("e.g.","").replace("i.e.","")
                # delete the label (can be extended if meet new format)
                review = re.sub('[1-9][\.)]'," ",review)
                review = re.sub('[^1-9a-zA-Z+-]'," ",review)
                review = review.lower()
                review = review.split()
                # delete stop words and transfer words to unified format
                review = [word for word in review if not word in set(stopwords.words("english"))]
                review = " ".join(review)
                corpus.append(review)
            # get cleaned discription
            basic_degree = corpus[0]
            basic_yoe = corpus[1]
            preferred_degree = corpus[2]
            preferred_yoe = corpus[3]
            final_output["basic_degree"] = self.get_degree_info(basic_degree)
            final_output["basic_major"] = self.get_major_info(basic_degree)
            final_output["preferred_degree"] = self.get_degree_info(preferred_degree)
            final_output["preferred_major"] = self.get_major_info(preferred_degree)
            final_output['basic_yoe'] = self.get_yoe_info(basic_yoe)
            final_output["preferred_yoe"] = self.get_yoe_info(preferred_yoe)
            return final_output

    def clean_responsibility(self):
        s_responsibility = self.data[self.company][self.position][self.id]['Responsibilities']
        if s_responsibility == "None":
            return []
        else:
            # split the sentences and store in the list (can be used for deducting the classification of words)
            sentences = nltk.sent_tokenize(s_responsibility)
            # store cleaned sentences
            basic_responsibility = []
            # clean data
            for i in range(len(sentences)):
                review = sentences[i].replace("etc.","").replace("e.g.","").replace("i.e.","")
                # only keep letters, + and - in the original passage
                review = re.sub('[^a-zA-Z+-]'," ",review)
                review = review.replace(" one "," ").replace(" two "," ").replace(" three "," ").replace(" four "," ").replace(" five "," ").replace(" six "," ").replace(" seven "," ").replace(" eight "," ").replace(" nine "," ").replace(" ten "," ")
                review = review.lower()
                review = review.split()
                # delete stop words and transfer words to unified format
                review = [word for word in review if not word in set(stopwords.words("english"))]
                review = " ".join(review)
                doc = nlp(review)
                basic_responsibility.append(" ".join([token.lemma_ for token in doc]))
            # return a list that contains cleaned sentences
            return basic_responsibility

    def clean_basic_qualification(self):
        s_basic_qualification = self.data[self.company][self.position][self.id]['Basic_Qualifications']
        if s_basic_qualification == "None":
            return []
        else:
            # split the sentences and store in the list (can be used for deducting the classification of words)
            sentences = nltk.sent_tokenize(s_basic_qualification)
            # store cleaned sentences
            corpus = []
            # clean data
            for i in range(len(sentences)):
                review = sentences[i].replace("etc.","").replace("e.g.","").replace("i.e.","")
                # only keep letters, + and - in the original passage
                review = re.sub('[^a-zA-Z+-]'," ",review)
                review = review.replace(" one "," ").replace(" two "," ").replace(" three "," ").replace(" four "," ").replace(" five "," ").replace(" six "," ").replace(" seven "," ").replace(" eight "," ").replace(" nine "," ").replace(" ten "," ")
                review = review.lower()
                review = review.split()
                # delete stop words and transfer words to unified format
                review = [word for word in review if not word in set(stopwords.words("english"))]
                review = " ".join(review)
                doc = nlp(review)
                corpus.append(" ".join([token.lemma_ for token in doc]))
            # return a list that contains cleaned sentences
            return corpus
    
    def clean_prefer_qualification(self):
        s_prefer_qualification = self.data[self.company][self.position][self.id]['Preferred_Qualifications']
        if s_prefer_qualification == "None":
            return []
        else:
            # split the sentences and store in the list (can be used for deducting the classification of words)
            sentences = nltk.sent_tokenize(s_prefer_qualification)
            # store cleaned sentences
            prefer_qualification = []
            # clean data
            for i in range(len(sentences)):
                review = sentences[i].replace("etc.","").replace("e.g.","").replace("i.e.","")
                # only keep letters, + and - in the original passage
                review = re.sub('[^a-zA-Z+-]'," ",review)
                review = review.replace(" one "," ").replace(" two "," ").replace(" three "," ").replace(" four "," ").replace(" five "," ").replace(" six "," ").replace(" seven "," ").replace(" eight "," ").replace(" nine "," ").replace(" ten "," ")
                review = review.lower()
                review = review.split()
                # delete stop words and transfer words to unified format
                review = [word for word in review if not word in set(stopwords.words("english"))]
                review = " ".join(review)
                doc = nlp(review)
                prefer_qualification.append(" ".join([token.lemma_ for token in doc]))
            # return a list that contains cleaned sentences
            return prefer_qualification

    def int_store_info(self):
        inner_info = {"Position_Country&City":self.data[self.company][self.position][self.id]["Position_Country&City"], 
                    "Recruiter":self.data[self.company][self.position][self.id]["Recruiter"], 
                    "Type":self.data[self.company][self.position][self.id]["Type"], 
                    "Seniority":self.data[self.company][self.position][self.id]["Seniority"], 
                    "Work_autho":self.data[self.company][self.position][self.id]["Work_autho"],
                    "Degree&YOE_Requirement": self.clean_degree_and_yoe(),
                    "Responsibilities":self.clean_responsibility(),
                    "Basic_Qualifications":self.clean_basic_qualification(),
                    "Preferred_Qualifications":self.clean_prefer_qualification(),
                    "Recruiting_History":self.data[self.company][self.position][self.id]["Recruiting_History"]}
        cleaned_info = {self.company:{self.position:{self.id:inner_info}}}
        # this step can be changed to other manipulation that can do storage
        return cleaned_info
```
