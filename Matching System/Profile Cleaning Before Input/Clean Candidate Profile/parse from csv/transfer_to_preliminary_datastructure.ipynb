```python
# as a class
import pandas as pd
import json
from datetime import date


class ParseFromEXCEL():
    
    # Note: Excel should be placed at the same location.
    def __init__(self,basic_info_xlsx,education_xlsx,work_exp_xlsx):
        self.basic_info_xlsx = basic_info_xlsx
        self.education_xlsx = education_xlsx
        self.work_exp_xlsx = work_exp_xlsx

    def transferDF(self):
        df1 = pd.read_excel(self.basic_info_xlsx)
        df1.columns = df1.iloc[0]
        df1.drop(df1.index[0],inplace=True)
        df2 = pd.read_excel(self.education_xlsx)
        df2.columns = df2.iloc[0]
        df2.drop(df2.index[0],inplace=True)
        df3 = pd.read_excel(self.work_exp_xlsx)
        df3.columns = df3.iloc[0]
        df3.drop(df3.index[0],inplace=True)
        return df1, df2, df3

    def buildFrame(self):
        final_dict = {}
        df1 = self.transferDF()[0]
        for id,name,code in zip(df1['id'],df1["name"],df1['area_code']):
            final_dict[id] = {
                "name":name.lower(),
                "Country&City":code,
                "Work_Autho":"None",
                "Degree&Major":"None",
                "YOE&Industry":"None",
                "skills":"None",
                "working_experiences":"None",
                "project_experiences":"None",
                "posting_history":[]
                }
        return final_dict

    def fillEducation(self):
        final_dict = self.buildFrame()
        df2 = self.transferDF()[1]
        for id in final_dict.keys():
            degree_major = []
            for degree, major, university in zip(df2[df2["canid_id"]==id]['degree'],df2[df2["canid_id"]==id]['major'],df2[df2["canid_id"]==id]['university']):
                if pd.isna(degree) or pd.isna(major):
                    continue
                elif pd.isna(university):
                    degree_major.append(f"{degree.lower()} {major.lower()}")
                else:
                    degree_major.append(f"{university.lower()} {degree.lower()} {major.lower()}")
            if len(degree_major)==0:
                final_dict[id]['Degree&Major'] = "None"
            else:
                final_dict[id]['Degree&Major'] = "|".join(degree_major)
        return final_dict
    
    def fillYOE(self):
        final_dict = self.fillEducation()
        df3 = self.transferDF()[2]
        for id in final_dict.keys():
            work_exp = []
            for company,position,start_date,end_date in zip(df3[df3["canid_id"]==id]['company'],df3[df3["canid_id"]==id]['position'],df3[df3["canid_id"]==id]["start_date"],df3[df3["canid_id"]==id]["end_date"]):
                if pd.isna(position) or pd.isna(start_date) or pd.isna(end_date):
                    continue
                elif pd.isna(company):
                    work_exp.append(f"{position.lower()} {start_date} {end_date}")
                else:
                    work_exp.append(f"{company.lower()} {position.lower()} {start_date} {end_date}")
            if len(work_exp)==0:
                final_dict[id]['YOE&Industry'] = "None"
            else:
                final_dict[id]['YOE&Industry'] = "|".join(work_exp)
        return final_dict

    def fillWorkExp(self):
        final_dict = self.fillYOE()
        df3 = self.transferDF()[2]
        for id in final_dict.keys():
            work_content = []
            for jd in df3[df3["canid_id"]==id]['job_description']:
                if pd.isna(jd):
                    continue
                else:
                    work_content.append(f"{jd.lower()}")
            if len(work_content)==0:
                final_dict[id]['working_experiences'] = "None"
            else:
                final_dict[id]['working_experiences'] = "|".join(work_content)
        return final_dict

    def Store_Info(self):
        today = str(date.today())
        final_dict = self.fillWorkExp()
        info_name = today+"information.json"
        with open(info_name,"w") as f:
            json.dump(final_dict,f)
```
