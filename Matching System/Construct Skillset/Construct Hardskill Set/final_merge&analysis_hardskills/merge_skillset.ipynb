```python
import pandas as pd
```

## Merge Datasets

```python
df1 = pd.read_excel("active_listening_skillset.xlsx")
df2 = pd.read_excel("math_skillset.xlsx")
df3 = pd.read_excel("speaking_skillset.xlsx")
df4 = pd.read_excel("writing_skillset.xlsx")
df5 = pd.read_excel("reading_comprehension_skillset.xlsx")
df6 = pd.read_excel("science_skillset.xlsx")
```

```python
df_final = df1.append(df2,ignore_index=True).append(df3,ignore_index=True).append(df4,ignore_index=True).append(df5,ignore_index=True).append(df6,ignore_index=True)
```

## Explore Final Datasets

```python
df_final.drop(columns='Unnamed: 0',inplace=True)
df_final.head()
```

```python
df_final.shape
```

```python
df_final.groupby(df_final['skill_name']).count().sort_values(by="skill_set",ascending=False)
```

```python
skill_name = list(set(list(df_final['skill_name'])))
len(skill_name)
```

```python
df_final.groupby([df_final['skill_name'],df_final['skill_set']]).count().sort_values(by="id",ascending=False)
```

```python
df_final[df_final['skill_name']=="Spreadsheet software"].groupby(df_final['skill_set']).count().sort_values(by="id",ascending=False)
```

## Merge On Skill_name Basis

```python
merged_skillset_list = []
for single_skill_name in skill_name:
    non_rep_skill = []
    df_temp = df_final[df_final['skill_name']==single_skill_name]
    for single_skill in df_temp['skill_set']:
        non_rep_skill = list(set(non_rep_skill+single_skill.strip("[").strip("]").split(",")))
    non_rep_skill = [i for i in non_rep_skill if i!=""]
    non_rep_skill_wopunc = [i.strip().strip("'") for i in non_rep_skill]
    merged_skillset_list.append(non_rep_skill_wopunc)
```

```python
non_rep_df = pd.DataFrame({"skill_name":skill_name,"skill_set":merged_skillset_list})
non_rep_df.head(5)
non_rep_df.tail(5)
```
