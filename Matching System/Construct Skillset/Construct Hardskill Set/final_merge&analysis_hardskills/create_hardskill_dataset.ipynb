```python
import pandas as pd
```

```python
df = pd.read_excel("cleaned_skillset.xlsx")
df.head()
```

```python
# delete the repetitive items and empty items
df.drop(columns='Unnamed: 0',inplace=True)
```

```python
# transfer skillset column to list
skill_set_list = [i.strip("[").rstrip("]").split(",") for i in df['skill_set']]
total_l = []
for l in skill_set_list:
    single_l = []
    for ele in l:
        single_l.append(ele.strip("[").strip("'").strip("'").strip(" '"))
    total_l.append(single_l)
# replace the original column
df['skill_set'] = total_l
```

```python
# create a dictionary
skill_dict = {}
for i in range(df.shape[0]):
    for m in df['skill_set'][i]:
        if m in skill_dict:
            skill_dict[m].append(df['skill_name'][i])
        else:
            skill_dict[m] = [df['skill_name'][i]]

# just for test
skill_dict['Java']
```

```python
patterns = []
for (skill_ele, skill_name) in skill_dict.items():
    patterns.append({"label": skill_name[0], "pattern": skill_ele.lower()})

patterns
```

```python
import json
with open("hardskills.json","w") as f:
    json.dump(patterns,f)
```
