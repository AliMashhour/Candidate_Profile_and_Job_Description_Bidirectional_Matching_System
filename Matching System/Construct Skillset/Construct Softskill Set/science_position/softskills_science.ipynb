```python
import pandas as pd
from lxml import etree
from selenium import webdriver

import time
```

```python
# escape the detection of chrome and set headless chrome
options = webdriver.ChromeOptions()
options.add_experimental_option("useAutomationExtension", False)
options.add_experimental_option("excludeSwitches",["enable-automation"])
options.add_argument("--headless")
options.add_argument("--disable-gpu")

web = webdriver.Chrome(chrome_options=options)
```

```python
df = pd.read_excel("science_job_title.xlsx")
df.head()
```

```python
url_base = "https://www.onetonline.org/link/summary/"
```

```python
content_name_list = []
detailed_description_list = []
for id in df['id']:
    url = url_base+id.strip()
    web.get(url)
    text = web.page_source
    tree = etree.HTML(text)
    try:
        work_content = tree.xpath("/html/body/div/div[1]/div/div[2]/div[8]/div[1]/div/ul/li")
        # /html/body/div/div[1]/div/div[2]/div[8]/div[1]/div/ul/li[1]
        content_set = {}
        for each_content in work_content:
            # /html/body/div/div[1]/div/div[2]/div[8]/div[1]/div/ul/li[1]/div/div[1]/b
            content_name = each_content.xpath("./div/div[1]/b/text()")[0]
            content_name_list.append(content_name.lower())
            # /html/body/div/div[1]/div/div[2]/div[8]/div[1]/div/ul/li[1]/div/div[1]/text()
            detailed_description = each_content.xpath("./div/div[1]/text()")[0]
            detailed_description_list.append(detailed_description.lower().lstrip(" - ").rstrip("."))
        print("finish"+id)
    except:
        print(id)
```
