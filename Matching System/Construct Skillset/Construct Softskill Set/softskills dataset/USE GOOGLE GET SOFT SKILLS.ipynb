```python
total_list = []
for i in df['KeyWords']:
    temp_list = []
    soft_skills = i.split()
    for core in soft_skills:
        if core != "-":
            temp_list = temp_list + [i[0].lower() for i in word_vectors.most_similar(positive=[core.lower().strip("-")],topn=15)]
    total_list.append(temp_list)

total_list
```

```python
## another method: calculate how many times each element appears. and rank them according to the number of times they appear. we will give some words higher priority if they appear more than others.
```

```python
# delete rep items
total_list_1 = []
for ele in total_list:
    total_list_1.append(list(set(ele)))

total_list_1
```

```python
# clean the items
total_list_2 = []
for ele in total_list_1:
    temp_list = []
    for item in ele:
        if ".com" in item or "#" in item or "." in item or "@" in item:
            continue
        else:
            temp_list.append(item.replace("_"," "))
    total_list_2.append(temp_list)

total_list_2
```

```python
import spacy

# English pipelines include a rule-based lemmatizer
nlp = spacy.load("en_core_web_sm")
lemmatizer = nlp.get_pipe("lemmatizer")
```

```python
total_list_3 = []
for ele in total_list_2:
    temp_list = []
    for item in ele:
        doc = nlp(item)
        temp_list.append(" ".join([token.lemma_ for token in doc]))
    total_list_3.append(temp_list)

total_list_3
```

```python
# delete rep items
total_list_4 = []
for ele in total_list_3:
    total_list_4.append(list(set(ele)))

total_list_4
```

```python
df_final = pd.DataFrame({"Target":list(df["Category"]),"KeyWords":total_list_4})
df_final.to_excel("final_skill_keyword.xlsx")
```

### explore final_skill_keyword

```python
explore_list = []
for i in total_list_4:
    explore_list = explore_list+i

len(explore_list)
```

```python
df_explore = pd.DataFrame({"KeyWord":explore_list})
temp_count = df_explore['KeyWord'].groupby(df_explore["KeyWord"]).count().sort_values(ascending=False)
```

```python
temp_dict = {}
flag = 0
for m,n in zip(list(temp_count.index),list(temp_count)):
    if n>6:
        temp_dict[m] = n
        flag = flag+1
    else:
        continue
print(flag)
temp_dict
```

### further cleaning with exploration

```python
total_list_5 = []
for i in total_list_4:
    temp_list = []
    for item in i:
        if item in list(temp_dict.keys()):
            continue
        else:
            temp_list.append(item)
    total_list_5.append(temp_list)

total_list_5
```

```python
df_final = pd.DataFrame({"Target":list(df["Category"]),"KeyWords":total_list_5})
df_final.to_excel("final_skill_keyword.xlsx")
```

```python
final_dict = {}
for i,p in zip(total_list_5,list(df["Category"])):
    for element in i:
        if element in final_dict.keys():
            final_dict[element].append(p)
        else:
            final_dict[element] = [p]
final_dict
```

```python
import json
with open("softskills.json","w") as f:
    json.dump(final_dict,f)
```
